```python
import pandas as pd
import numpy as np
data = pd.read_csv("/content/Book2.csv")

data.head()
```

```python
concepts = np.array(data)[:,:-1]

print("Concepts :-",concepts+)
target = np.array(data)[:,-1]

print("Target:-",target)

```

```python
def train(con,tar):
  for i,val in enumerate(tar):
    if(val == 'yes'):
      specific_h = con[i].copy()
      print(f"\n Initial hypothesis from first positive example :- {specific_h}")
      break
  for i,val in enumerate(con):
    if(tar[i] == 'yes'):
      for x in range(len(specific_h)):
        if val[x] != specific_h[x]:
          specific_h[x]='?'
      print(f"\n hypothesis after example {i+1}: {specific_h}")
  return specific_h
specific_h = train(concepts,target)
print(f"\n Final hypothesis :- {specific_h}")
```
